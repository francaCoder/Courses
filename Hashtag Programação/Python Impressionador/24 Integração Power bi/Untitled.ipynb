```python
import pandas as pd

sales_df = pd.read_csv(r"C:\Users\mathe\PycharmProjects\pythonProgram\pythonCodes\Python Impressionador\24 Integração Power bi\Contoso - Vendas  - 2017.csv", sep=";")
products_df = pd.read_csv(r"C:\Users\mathe\PycharmProjects\pythonProgram\pythonCodes\Python Impressionador\24 Integração Power bi\Contoso - Cadastro Produtos.csv", sep=";")
store_df = pd.read_csv(r"C:\Users\mathe\PycharmProjects\pythonProgram\pythonCodes\Python Impressionador\24 Integração Power bi\Contoso - Lojas.csv", sep=";")
clients_df = pd.read_csv(r"C:\Users\mathe\PycharmProjects\pythonProgram\pythonCodes\Python Impressionador\24 Integração Power bi\Contoso - Clientes.csv", sep=";")
                         
# Filter
# print(products_df.info())
clients_df = clients_df[["ID Cliente", "E-mail"]]
products_df = products_df[["ID Produto", "Nome da Marca"]]
store_df = store_df[["ID Loja", "Nome da Loja"]]

#
sales_df = sales_df.merge(products_df, on='ID Produto')
sales_df = sales_df.merge(store_df, on='ID Loja')
sales_df = sales_df.merge(clients_df, on='ID Cliente').rename(columns={"E-mail": "E-mail do Cliente"})
display(sales_df)
```

```python
dataset = sales_df
tree_stores_df = dataset[dataset['ID Loja'].isin([86, 172, 306])]
display(tree_stores_df)
```

```python
import matplotlib.pyplot as plt
tree_stores_df.plot(x='Data da Venda', y='Quantidade Vendida', figsize=(15, 5))
plt.show()
```

```python
import seaborn as sns

# sns.set_theme(style="darkgrid")

sns.lineplot(x='Data da Venda', y='Quantidade Vendida', hue='Nome da Loja', data=tree_stores_df)

plt.show()
```
